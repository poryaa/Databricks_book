```python
import os
import json
import shutil

# Read credentials from kaggle.json file
kaggle_source = "/Workspace/Users/paminidigehsara@rosen-nxt.com/Book_Chapter02/kaggle.json"
with open(kaggle_source) as f:
    creds = json.load(f)

# Set environment variables for Kaggle API
os.environ['KAGGLE_USERNAME'] = creds['username']
os.environ['KAGGLE_KEY'] = creds['key']

# Copy to ~/.kaggle/ so opendatasets doesn't prompt interactively
kaggle_dir = os.path.expanduser("~/.kaggle")
os.makedirs(kaggle_dir, exist_ok=True)
shutil.copy(kaggle_source, os.path.join(kaggle_dir, "kaggle.json"))
os.chmod(os.path.join(kaggle_dir, "kaggle.json"), 0o600)

print(f"Kaggle credentials loaded for user: {creds['username']}")
```

```python
#Create raw data directory
raw_data_path = volume_file_path + 'raw_data'
dbutils.fs.mkdirs(raw_data_path)

#Download data using kaggle API directly (avoids opendatasets interactive prompt)
import kaggle
kaggle.api.username = os.environ['KAGGLE_USERNAME']
kaggle.api.key = os.environ['KAGGLE_KEY']
kaggle.api.authenticate()
kaggle.api.competition_download_files("store-sales-time-series-forecasting", path=raw_data_path, quiet=False)

# Extract the downloaded zip
import zipfile
zip_path = os.path.join(raw_data_path, "store-sales-time-series-forecasting.zip")
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(os.path.join(raw_data_path, "store-sales-time-series-forecasting"))

dbutils.fs.ls(raw_data_path + "/store-sales-time-series-forecasting/")
```
