```python
import seaborn as sns
g = sns.PairGrid(spark.table('porya_catalog.default.synthetic_transactions').toPandas()[['CustomerID','Amount', 'Product','Label']], diag_sharey=False, hue="Label")
g.map_lower(sns.kdeplot).map_diag(sns.kdeplot, lw=3).map_upper(sns.regplot).add_legend()
```
